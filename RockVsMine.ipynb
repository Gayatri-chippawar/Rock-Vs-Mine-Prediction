```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('../datasets/sonar_data.csv',header=None)
```

```python
data.head()
```

```python
data.tail()
```

```python
data[60].value_counts()
```

```python
data.shape
```

```python
x = data.drop(columns=60,axis=1)
y = data[60]
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,stratify=y,random_state=3)
```

```python
print(x_train.shape,x_test.shape,x.shape)
```

```python
print(y_train.shape,y_test.shape,y.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model_predict_train = model.predict(x_train)
train_accuracy = accuracy_score(y_train,model_predict_train)
```

```python
print(train_accuracy)
```

```python
model_predict_test = model.predict(x_test)
test_accuracy = accuracy_score(y_test,model_predict_test)
```

```python
print(test_accuracy)
```
